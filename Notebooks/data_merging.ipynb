```python
import zipfile
import pandas as pd
import logging
import os

# Log fayl manzili
log_path = r"C:\Users\Rasulbek907\Desktop\SML_2_Pr\Log\data_loader.log"

# Log sozlamalari
logging.basicConfig(
    filename=log_path,
    filemode='a',  
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

def load_csv_from_zip(zip_path, csv_filename):
    try:
        logging.info(f"ZIP fayl ochilmoqda: {zip_path}")
        with zipfile.ZipFile(zip_path) as z:
            with z.open(csv_filename) as f:
                df = pd.read_csv(f)
        logging.info(f"'{csv_filename}' muvaffaqiyatli o'qildi.")
        return df
    except Exception as e:
        logging.error(f"Xatolik: {zip_path} dan '{csv_filename}' ni o'qishda: {e}")
        return None

if __name__ == "__main__":
    zip_train = r"C:\Users\Rasulbek907\Desktop\SML_2_Pr\Data\raw_data\train.csv.zip"
    zip_test = r"C:\Users\Rasulbek907\Desktop\SML_2_Pr\Data\raw_data\test.csv.zip"
    
    train_csv = "train.csv"
    test_csv = "test.csv"
    
    logging.info("Ma'lumotlarni yuklash jarayoni boshlandi.")
    
    train_df = load_csv_from_zip(zip_train, train_csv)
    test_df = load_csv_from_zip(zip_test, test_csv)
    
    if train_df is not None:
        logging.info(f"Train dataset satrlar soni: {len(train_df)}")
    else:
        logging.warning("Train dataset yuklanmadi.")
    
    if test_df is not None:
        logging.info(f"Test dataset satrlar soni: {len(test_df)}")
    else:
        logging.warning("Test dataset yuklanmadi.")
    
    logging.info("Ma'lumotlarni yuklash jarayoni tugadi.")
```
